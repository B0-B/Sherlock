```python
import socket
import ipaddress

def get_local_ips(network='192.168.1.0/24'):
    ip_net = ipaddress.ip_network(network)
    local_ips = []

    for ip in ip_net:
        sock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
        sock.settimeout(0.1)
        result = sock.connect_ex((str(ip), 80))
        if result == 0:
            print(str(ip))
            local_ips.append(str(ip))
        sock.close()

get_local_ips()
```

```python
import psutil

# Get all network interfaces and their IP addresses
def get_interfaces_and_ips():
    interfaces = psutil.net_if_addrs()
    result = {}
    for interface_name, interface_addresses in interfaces.items():
        print('interface', interface_name)
        for address in interface_addresses:
            print('\t', 'obj', address)
            print('\t', 'address', address.address)
            print('\t', 'family', address.family)
            print('\t', 'netmask', address.netmask, '\n')
    return result

if __name__ == "__main__":
    interfaces_and_ips = get_interfaces_and_ips()
```

```python
import psutil, socket

def get_mac_addresses () -> list[str]:

    interfaces = psutil.net_if_addrs()
    
    for interface_addresses in interfaces.values():

        for address in interface_addresses:

            if address.family == psutil.AF_LINK:
                print('MAC', address.address)
            elif address.family == socket.AF_INET:
                print('IP', address.address)

get_mac_addresses()
```

```python
import scapy.all as scapy

def scan(ip_range):
    print(f"Scanning IP range: {ip_range}")
    arp_request = scapy.ARP(pdst=ip_range)
    broadcast = scapy.Ether(dst="ff:ff:ff:ff:ff:ff")
    arp_request_broadcast = broadcast / arp_request
    print("Sending ARP requests...")
    answered_list = scapy.srp(arp_request_broadcast, timeout=5, verbose=True)[0]
    if not answered_list:
        print("No responses received.")
    else:
        print("Responses received.")
        devices = []
        for element in answered_list:
            device = {'ip': element[1].psrc, 'mac': element[1].hwsrc}
            devices.append(device)
        print(f"Device found: IP = {device['ip']}, MAC = {device['mac']}")
        return devices

# Example usage
devices = scan("192.168.1.0/24")
for device in devices:
    print(f"IP: {device['ip']}, MAC: {device['mac']}")
```

```python
import os 

def scan () -> list[str]:

    # Parameters
    timeout = 50
    count = 4
    
    # Determine host address.
    ip = get_host_ip()

    # From https://en.wikipedia.org/wiki/IPv4:
    # Private networks run usually from 192.168.0.0 - 192.168.255.255 accounting for 65k addresses.
    # 

    # Determine the subnet part.
    subnet = '.'.join(ip.split('.')[:3])
    ips = []


    for host_num in range(256):
        
        # Construct ip for scanning
        ip = f"{subnet}.{host_num}" 
    
        # Run single packet ping
        response = os.system(command = f"ping -n {count} -w {timeout} {ip}")

        # Check the response
        if response == 0:
            ips.append(ip)
            print(f"Ping to {ip} successful!")
        else:
            print(f"Ping to {ip} failed.")
    
    return ips

scan()
```

```python
import subprocess
import re

# Function to get MAC address for a given IP
def get_mac(ip) -> str|None:

    try:

        # Execute the arp command
        output = subprocess.check_output(['arp', '-a', ip]).decode('utf-8')

        # Use regex to find the MAC address, group to a string and transform to MAC standard.
        mac_address = ':'.join( (re.search(r'([a-fA-F0-9]{2}[:-]){5}[a-fA-F0-9]{2}', output).group()).split('-') )

        return mac_address
    except Exception as e:
        print(e)
        return 

get_mac('192.168.178.34')
```

```python
import subprocess

class NetworkScanner:

    '''
    A private network scanner which scans all sub addresses in the 
    local host's subnet address range, and updates the instance ips list. 
    '''

    def __init__(self):

        # Parameters
        self.timeout = 50
        self.count = 4

        self.ips: list[str] = []
        
        # Determine the host and subnet part
        self.host: str = get_host_ip()
        self.subnet: str = '.'.join(self.host.split('.')[:3])
    
    def __call__(self, num_start: int=0, num_stop: int=255) -> list[str]:

        '''
        Alias for scan_range.
        '''

        return self.scan_range(num_start, num_stop)

    def scan_range (self, num_start: int=0, num_stop: int=255) -> list[str]:

        '''
        Scans all sub addresses in the local hosts subnet address range,
        and overrides the instance ips list. 
        Example: if the host is 192.168.178.24 -> 192.168.178 will be the subnet.
        The scan will run in a range 192.168.178.num_start - 192.168.178.num_stop.

        [Return]
        
        Returns the ips list.
        '''

        pings = {}

        for num in range(num_start, num_stop):

            ip = f'{self.subnet}.{num}'
            process = subprocess.Popen(['ping', '-n', str(self.count), '-w', str(self.timeout), ip])
            pings[ip] = process
        
        # Wait for all processes to complete and collect their output
        active_ips = []
        for ip, process in pings.items():
            process.wait()  # Wait for the process to complete
            # output, error = process.communicate()  # Get the output and error from the process
            if process.returncode == 0:
                # print(f"Ping to {ip} successful!")
                active_ips.append(ip)
            # else:
            #     pings[ip] = False
                # print(f"Ping to {ip} failed.")
        
        # Override the ips object
        self.ips = active_ips

        return active_ips
    
NetworkScanner().scan_range(0, 255)
```
